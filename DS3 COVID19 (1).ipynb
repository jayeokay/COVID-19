```python
import pandas as pd
```

```python
Comorbidity_data = pd.read_csv('Conditions_contributing_to_deaths_involving_coronavirus_disease_2019__COVID-19___by_age_group_and_state__United_States..csv')
```

```python
Comorbidity_data
```

```python
Comorbidity_data.sort_values(by = 'Condition')
US = Comorbidity_data.loc[(Comorbidity_data['State']=='CA') | (Comorbidity_data['State']=='CO')
                         | (Comorbidity_data['State']=='CT')| (Comorbidity_data['State']=='GA')
                          | (Comorbidity_data['State']=='MD')| (Comorbidity_data['State']=='MN')
                          | (Comorbidity_data['State']=='NM')| (Comorbidity_data['State']=='NY')
                          | (Comorbidity_data['State']=='OR')| (Comorbidity_data['State']=='TN')
                    
                         ]
respiratory = US.loc[US['Condition Group'] == 'Respiratory diseases']
respiratory_all = respiratory.loc[respiratory['Age Group'] == 'All Ages']
respiratory_all = respiratory_all.reset_index()
respiratory_all
```

```python
sum_lung = 0
for i in range(60):
    sum_lung = sum_lung + respiratory_all.iloc[i, 9]

sum_lung
```

```python
C = US.loc[US['Condition Group'] == 'Circulatory diseases']
C_all = C.loc[C['Age Group'] == 'All Ages']
C_all = C_all.reset_index()
C_all
```

```python
sum_C = 0
for i in range(70):
    sum_C = sum_C + int(C_all.iloc[i, 9])

sum_C = sum_C
sum_C
```

```python
S = US.loc[US['Condition Group'] == 'Sepsis']
S_all = S.loc[S['Age Group'] == 'All Ages']
S_all = S_all.reset_index()

S_all
```

```python
sum_S = 0
for i in range(10):
    sum_S = sum_S + S_all.iloc[i, 9]

sum_S
```

```python
O = US.loc[US['Condition Group'] == 'Obesity']
O_all = O.loc[O['Age Group'] == 'All Ages']
O_all = O_all.reset_index()

O_all
```

```python
sum_O = 0
for i in range(10):
    sum_O = sum_O + O_all.iloc[i, 9]

sum_O
```

```python
ND = US.loc[US['Condition Group'] == 'Alzheimer disease']
ND_all = ND.loc[ND['Age Group'] == 'All Ages']
ND_all = ND_all.reset_index()

ND1 = US.loc[US['Condition Group'] == 'Vascular and unspecified dementia']
ND1_all = ND1.loc[ND1['Age Group'] == 'All Ages']
ND1_all = ND1_all.reset_index()

print(ND_all, ND1_all)
```

```python
sum_ND = 0
for i in range(10):
    sum_ND = sum_ND + ND_all.iloc[i, 9]

sum_ND1 = 0
for i in range(10):
    sum_ND1 = sum_ND1 + ND1_all.iloc[i, 9]

sum_ND = sum_ND + sum_ND1
sum_ND
```

```python
R = US.loc[US['Condition Group'] == 'Renal failure']
R_all = R.loc[R['Age Group'] == 'All Ages']
R_all = R_all.reset_index()
R_all
```

```python
sum_R = 0
for i in range(10):
    sum_R = sum_R + R_all.iloc[i, 9]
    
sum_R
```

```python
import matplotlib.pyplot as plt

medical_conditions = ['Respiratory', 'Circulatory', 'Immune related', 'Obesity', 'Neurological', 'Renal']
no_cases = [sum_lung, sum_C, sum_S, sum_O, sum_ND, sum_R]

plt.figure(figsize=(9, 15))
plt.bar(medical_conditions, no_cases, label='All age groups')
plt.legend()
plt.xlabel('Conditions')
plt.ylabel('Number of Deaths')
```
