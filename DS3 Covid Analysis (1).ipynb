```python
import requests
import urllib.request
import babypandas as bpd
import numpy as np
import csv
from bs4 import BeautifulSoup
```

```python
url= "https://care.diabetesjournals.org/content/43/7/1399.figures-only"
response = requests.get(url)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, "html.parser")

```

```python
table = soup.findAll('div', {"class":"table-expansion"})[0]
table
values = []
tr = table.findAll(['tr'])[0:35]
```

```python
csvFile = open("table1.csv", 'wt', newline='', encoding='utf-8')
writer= csv.writer(csvFile)
try:
    for cell in tr:
        th = cell.find_all('th')
        th_data = [col.text.strip('\n') for col in th]
        td = cell.find_all('td')
        row = [i.text.replace('\n', '') for i in td]
        writer.writerow(th_data+row)
finally:
    csvFile.close()
    
    
```

```python
demographics = bpd.read_csv('table1.csv')
demographics
```

```python
age_info = demographics.loc[0:4]
age_info = age_info.assign(Age_Group=age_info.get("Unnamed: 0")).drop(columns="Unnamed: 0").set_index("Age_Group").reset_index().loc[1:4]
age_info = age_info.drop(columns="Patients with clinically diagnosed COVID-19 (n = 563)")
age_info

```

```python
sex_info = demographics.loc[5:7]
sex_info = sex_info.assign(Patient_Sex=sex_info.get("Unnamed: 0")).drop(columns="Unnamed: 0").set_index("Patient_Sex").reset_index().loc[1:3]
sex_info = sex_info.drop(columns="Patients with clinically diagnosed COVID-19 (n = 563)")
sex_info
```

```python
symptoms_info = demographics.loc[8:16]
symptoms_info = symptoms_info.assign(Initial_Symptoms=symptoms_info.get("Unnamed: 0")).drop(columns="Unnamed: 0").set_index("Initial_Symptoms").reset_index().loc[1:9]
symptoms_info = symptoms_info.drop(columns="Patients with clinically diagnosed COVID-19 (n = 563)")
symptoms_info

```

```python
comorbidities = demographics.loc[17:24]
comorbidities = comorbidities.assign(Preexisting_Conditions=comorbidities.get("Unnamed: 0")).drop(columns="Unnamed: 0").set_index("Preexisting_Conditions").reset_index().loc[1:8]
comorbidities = comorbidities.drop(columns="Patients with clinically diagnosed COVID-19 (n = 563)")
comorbidities
```

```python
covid_severity_info = demographics.loc[25:28]
covid_severity_info = covid_severity_info.assign(Initial_Disease_Severity=covid_severity_info.get("Unnamed: 0")).drop(columns=["Unnamed: 0"]).set_index("Initial_Disease_Severity").reset_index().loc[1:3]
covid_severity_info = covid_severity_info.drop(columns="Patients with clinically diagnosed COVID-19 (n = 563)")
covid_severity_info
```
